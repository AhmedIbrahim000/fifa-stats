```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the data from the CSV file
df_clus = pd.read_csv('Fifa_stats_cleaning.csv')

df_clus['Position'] = df_clus['Position'].replace({
    'CAM':'5',
    'CDM':'5',
    'CM':'5',
    'CB':'4',
    'LB':'3',
    'RB':'2',
    'CF':'10',
    'ST':'9',
    'GK':'1',
    'LCB':'4',
    'LCM':'5',
    'LDM':'5',
    'LAM':'5',
    'LF':'10',
    'LM':'5',
    'LS':'9',
    'LW':'11',
    'LWB':'3',
    'RAM':'5',
    'RCB':'2',
    'RCM':'5',
    'RDM':'5',
    'RF':'10',
    'RM':'5',
    'RS':'9',
    'RW':'11',
    'RWB':'2'
})
```

```python
# Select the features to be used for clustering
X = df_clus[['Overall', 'Potential', 'Position', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing',
       'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing',
       'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions',
       'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']]
```

```python
# Select the features to be used for clustering
X1 = df_clus[['LongShots', 'Curve']]
```

```python
X2 = df_clus[['Finishing', 'ShotPower']]
```

```python
# Instantiate and fit the K-means model
kmeans_1 = KMeans(n_clusters=2, random_state=42)
kmeans_1.fit(X1)

# Predict the cluster labels
labels_1 = kmeans_1.predict(X1)

# Get the coordinates of the cluster centers
centers_1 = kmeans_1.cluster_centers_

# SAVING THE PREDECTED VALUES
df_pred_kmeans = pd.DataFrame(labels_1, columns=["PredictedValue"])
df_pred_kmeans.to_csv("K-means(L,C).csv", index=False)
```

```python
# Instantiate and fit the K-means model
kmeans_2 = KMeans(n_clusters=2, random_state=42)
kmeans_2.fit(X2)

# Predict the cluster labels
labels_2 = kmeans_2.predict(X2)

# Get the coordinates of the cluster centers
centers_2 = kmeans_2.cluster_centers_

# SAVING THE PREDECTED VALUES
df_pred_kmeans = pd.DataFrame(labels_1, columns=["PredictedValue"])
df_pred_kmeans.to_csv("K-means(F,S).csv", index=False)
```

```python
# Create a scatter plot of the data points
plt.scatter(X1['LongShots'], X1['Curve'], c=labels_1)
plt.scatter(centers_1[:, 0], centers_1[:, 1], c='red', marker='x')
plt.xlabel('LongShots')
plt.ylabel('Curve')
plt.title('K-means Clustering')
plt.show()
```

```python
# Visualize the clusters and cluster centers
plt.scatter(X2['Finishing'], X2['ShotPower'], c=labels_2)
plt.scatter(centers_2[:, 0], centers_2[:, 1], c='red', marker='x')
plt.xlabel('Finishing')
plt.ylabel('ShotPower')
plt.title('K-means Clustering')
plt.show()
```
