```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(rc={'figure.figsize': [15, 8]}, font_scale=1.2) # Standard figure size for all

df = pd.read_csv("fifa_eda_stats.csv", index_col=0, na_values=True)
```

```python
df.shape
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns = ['International Reputation','Jersey Number','Loaned From'], axis=1, inplace=True)
```

```python
df.info()
```

```python
column_to_check = df.columns
column_to_check
```

```python
for column in column_to_check:
    if df[column].dtype == float:
        median_value = df[column].loc[df[column].apply(lambda x: isinstance(x, (int, float)))].median()
        df[column].fillna(median_value, inplace=True)
```

```python
df.isna().sum()
```

```python
# Handling the object columns by fillna values with mode 9 columns
for col in df.columns:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
# cleaning column value and wage ,convert M,K with numeric value and convert the column Type
df['Value'] = df['Value'].apply(lambda x:x.split('€')[1])
```

```python
df['Value'] = df['Value'].replace({'M': '*1e6', 'K': '*1e3'}, regex=True).map(pd.eval)
```

```python
df['Value']
```

```python
df['Value'] = df['Value'].astype(int)
```

```python
df['Value']
```

```python
df['Wage'] = df['Wage'].apply(lambda x:x.split('€')[1])
```

```python
df['Wage'] = df['Wage'].replace({'K': '000', 'M': '000000'}, regex=True)
```

```python
df['Wage'] = pd.to_numeric(df['Wage'],errors='coerce')
```

```python
df['Wage'] = df['Wage'].astype(int)
```

```python
df['Wage']
```

```python
# Convert the Weight from pound to kilogram
df['Weight'] = df['Weight'].apply(lambda x:x.split('lbs')[0])
```

```python
df['Weight']
```

```python
df['Weight'] = pd.to_numeric(df['Weight'])
```

```python
df['Weight'] = df['Weight']*0.45359237
```

```python
df['Weight'] = df['Weight'].astype(int)
df['Weight']
```

```python
# Convert the Joined date to dateTime format
df['Joined']=pd.to_datetime(df['Joined'])
```

```python
df['joined_month'] = df['Joined'].dt.month
```

```python
df['Joined_year']=df['Joined'].dt.year
```

```python
df['Joined_day'] = df['Joined'].dt.day
```

```python
# handle height cleaning and convert fromm height by feet to height by meter
df['Height_n1'] = df['Height'].apply(lambda x:x.split("'")[0])
```

```python
df['Height_n2'] = df['Height'].apply(lambda x:x.split("'")[1])
```

```python
df['Height_n1']
```

```python
df['Height_n2']
```

```python
df['Height_n1'] = pd.to_numeric(df['Height_n1'], errors='coerce')
```

```python
df['Height_n2'] = pd.to_numeric(df['Height_n2'], errors='coerce')
```

```python
df['Height_m']=(df['Height_n1']* 0.3048)+(df['Height_n2']*0.0254)
```

```python
# handle release clause column
df['Release Clause']
```

```python
df['Release Clause'] = df['Release Clause'].apply(lambda x:x.split('€')[1])
```

```python
df['Release Clause']=df['Release Clause'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval)
```

```python
df['Release Clause'] = df['Release Clause'].astype(int)
```

```python
df['Release Clause']
```
